# Procedural Frozen Lake — random rollout

This notebook shows the standard Gymnasium `reset()` / `step()` loop with `Procedural-FrozenLake-v1`.

Each episode can use a procedurally generated map with multiple possible start/goal tiles and per-goal rewards. Pass `options={"regenerate_map": True}` to `reset()` to sample a fresh layout every episode.

```python
import gymnasium as gym
import procedural_frozenlake

env = gym.make(
    procedural_frozenlake.PROCEDURAL_FROZENLAKE_ENV_ID,
    map_seed=7,
    emit_map=True,
    emit_q_star=True,
    step_penalty=-0.01,
    hole_prob=0.15,
)
```

```python
obs, info = env.reset(seed=0, options={"regenerate_map": True})
print("Map:", info["map"])
print("Observation:", obs)
print("Q* for current state:", info["q_star"])
```

```python
total_reward = 0.0
for step in range(200):
    action = int(info["q_star"].argmax())  # greedy expert
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    if terminated or truncated:
        print(f"Episode ended at step {step + 1}, return={total_reward:.3f}")
        obs, info = env.reset(options={"regenerate_map": True})
        total_reward = 0.0
        break

env.close()
```
